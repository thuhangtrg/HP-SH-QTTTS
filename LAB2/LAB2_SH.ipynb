### Phần 4: Trực quan hóa dữ liệu nâng cao


```python
#câu 1: Vẽ biểu đồ đường Simple cho biến T1
import matplotlib.pyplot as plt
t1_counts = data['T1'].value_counts().sort_index()
plt.plot(t1_counts.index, t1_counts.values)
plt.title('Số lượng học sinh theo từng giá trị T1', fontsize=14)
plt.xlabel('Giá trị T1', fontsize=12)
plt.ylabel('Số lượng học sinh', fontsize=12)
plt.show()
```

```python
# 2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
# a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)
# b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)
# c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)
# d. Từ 8 trở lên = giỏi (ký hiệu “g”)

def classify_t1(x):
    if x < 5:
        return 'k'  # Kém
    elif x < 7:
        return 'tb'  # Trung bình
    elif x < 8:
        return 'k'  # Khá
    else:
        return 'g'  # Giỏi

# Áp dụng phân loại
data['phanlopt1'] = data['T1'].apply(classify_t1)

# Kiểm tra dữ liệu
print(data[['T1', 'phanlopt1']].head())  # Hiển thị 5 dòng đầu

```

```python
# 3. Lập bảng tần số cho biến phanloait1
# Lập bảng tần số cho biến phanlopt1
phanlopt1_freq = data['phanlopt1'].value_counts()
print(phanlopt1_freq)
```

```python
data['phanlopt1'] = pd.cut(
    data['T1'],
    bins=[0, 5, 7, 8, 10],
    labels=['Y', 'TB', 'K', 'G'],
    right=False
)

data[['T1', 'phanlopt1']].head()
phanlopt1_counts = data['phanlopt1'].value_counts()
print(phanlopt1_counts)

t1_grouped = data.groupby(['T1', 'phanlopt1']).size().unstack()
plt.figure(figsize=(8, 5))
t1_grouped.plot(kind='line', linestyle='-', linewidth=2)
plt.legend(title="Phân loại", labels=["Kém (k)", "Trung bình (tb)", "Khá (k)", "Giỏi (g)"])
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
t1_class_counts = data.groupby(['phanlopt1', 'T1']).size().unstack(fill_value=0)
for label in t1_class_counts.index:
    plt.scatter(t1_class_counts.columns, t1_class_counts.loc[label], label=label)
    for x, y in zip(t1_class_counts.columns, t1_class_counts.loc[label]):
        plt.plot([x, x], [0, y], linestyle='-', color='black', alpha=0.5)
plt.title('Biểu đồ Drop-line cho T1 theo phân loại phanlopt1', fontsize=14)
plt.xlabel('Giá trị T1', fontsize=12)
plt.ylabel('Số lượng học sinh', fontsize=12)

plt.legend(title='Phân loại T1')
plt.show()
```

Phần 5: Mô tả dữ liệu và khảo sát dạng phân phối

```python
# 1. Hãy mô tả và khảo sát phân phối cho biến T1
# Gợi ý
# - Mô tả độ tập trung và phân tán của dữ liệu T1
# - Vẽ biểu đồ Box-Plot và xác định các 10 đại lượng trong biểu đồ đó
# - Mô tả hình dáng lệch của phân phối T1 dựa vào các đại lượng hướng tâm
# - Vẽ biểu đồ Histogram biểu thị hình dáng phân phối
# - Mô tả các đặc trưng của phân phối, mức độ lệch và mức độ nhọn
# - Kiểm chứng phân phối chuẩn QQ-Plot
# - Nhận xét và đánh giá về phân phối của T1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Mô tả thống kê
print(data["T1"].describe())

# Vẽ Box-Plot
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.boxplot(x=data["T1"])
plt.title("Box-Plot của T1")

# Vẽ Histogram
plt.subplot(1, 3, 2)
sns.histplot(data["T1"], bins=10, kde=True)
plt.title("Histogram của T1")

# Vẽ QQ-Plot
plt.subplot(1, 3, 3)
stats.probplot(data["T1"], dist="norm", plot=plt)
plt.title("QQ-Plot của T1")

plt.tight_layout()
plt.show()
()
```

```python
# 2. Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
# Gợi ý
# - Trực quan hóa biểu đồ Box-plot, histogram và QQ-plot theo phân nhóm là giá trị
# của ‘phanlopT1’.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Vẽ Box-Plot theo nhóm phanlopT1
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.boxplot(x="phanlopt1", y="T1", data=data)
plt.title("Box-Plot của T1 theo phanlopT1")

# Vẽ Histogram theo nhóm phanlopT1
plt.subplot(1, 3, 2)
sns.histplot(data, x="T1", hue="phanlopt1", bins=10, kde=True, element="step")
plt.title("Histogram của T1 theo phanlopT1")

# Vẽ QQ-Plot theo nhóm phanlopT1
plt.subplot(1, 3, 3)
for category in data["phanlopt1"].unique():
    stats.probplot(data[data["phanlopt1"] == category]["T1"], dist="norm", plot=plt)
plt.title("QQ-Plot của T1 theo phanlopT1")

plt.tight_layout()
plt.show()
```

```python
# 3. Hãy khảo sát tương quan giữa biến DH1 theo biến T1
# Gợi ý
# - Nhận xét giá trị Covariance hoặc Correlation
# - Vẽ biểu đồ Scatter thể hiện liên hệ của biến phụ thuộc DH1 theo biến độc lập T1

import seaborn as sns
import matplotlib.pyplot as plt

# Tính Covariance và Correlation
cov_value = data["DH1"].cov(data["T1"])
corr_value = data["DH1"].corr(data["T1"])

print(f"Covariance giữa DH1 và T1: {cov_value:.2f}")
print(f"Correlation giữa DH1 và T1: {corr_value:.2f}")

# Vẽ biểu đồ Scatter
plt.figure(figsize=(6, 4))
sns.scatterplot(x=data["T1"], y=data["DH1"])
plt.xlabel("T1")
plt.ylabel("DH1")
plt.title("Biểu đồ Scatter giữa DH1 và T1")
plt.show()
```
